#  EDA for Churn Dataset

This notebook checks the basic data quality points needed for Part A:

- dataset shape and preview
- missing values per feature
- number of distinct values per feature
- contract type distribution
- churn target balance

```python
from pathlib import Path

import pandas as pd

DATA_PATH = Path("../data/customers.csv")

df = pd.read_csv(DATA_PATH)
df.head()
```

## Dataset Size

```python
rows, columns = df.shape
print(f"Rows: {rows}")
print(f"Columns: {columns}")
```

## Missing Values and Distinct Values per Feature

```python
feature_summary = pd.DataFrame({
    "feature": df.columns,
    "dtype": df.dtypes.astype(str).values,
    "missing_values": df.isna().sum().values,
    "distinct_values": df.nunique(dropna=False).values,
})

feature_summary
```

## Contract Type Distinct Values

```python
contract_type_counts = (
    df["contract_type"]
    .value_counts(dropna=False)
    .rename_axis("contract_type")
    .reset_index(name="count")
)

contract_type_counts["percent"] = (
    contract_type_counts["count"] / len(df) * 100
).round(2)

contract_type_counts
```

## Churn Target Balance

```python
target_balance = (
    df["churned"]
    .value_counts(dropna=False)
    .rename_axis("churned")
    .reset_index(name="count")
)

target_balance["percent"] = (
    target_balance["count"] / len(df) * 100
).round(2)

target_balance
```
